### Now let's get the parameters (ðœ‡ andðœŽ2) for a Gaussian distribution "mu" = ðœ‡, "var" = ðœŽ

```python
def estimate_gaussain(X):
    m, n = X.shape
    mu = 1/m * np.sum(X, axis=0)
    var = 1/m * np.sum((X - mu) **2, axis=0)
    
    return mu, var
```
